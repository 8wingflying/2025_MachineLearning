# 矩陣 SVD (奇異值分解) 推導與驗證

## 1. 問題定義

給定矩陣 $A$：

$$
A = \begin{bmatrix} 3 & 1 & 1 \\ -1 & 3 & 1 \end{bmatrix}
$$

目標是找出 $A = U \Sigma V^T$，其中：
* $U$ 是 $2 \times 2$ 的正交矩陣。
* $\Sigma$ 是 $2 \times 3$ 的對角矩陣（包含奇異值）。
* $V$ 是 $3 \times 3$ 的正交矩陣。

---

## 2. 逐步推導過程

### 第一步：計算 $AA^T$ 並求解 $\Sigma$ 與 $U$

首先計算 $AA^T$：

$$
AA^T = \begin{bmatrix} 3 & 1 & 1 \\ -1 & 3 & 1 \end{bmatrix} \begin{bmatrix} 3 & -1 \\ 1 & 3 \\ 1 & 1 \end{bmatrix} = \begin{bmatrix} 11 & 1 \\ 1 & 11 \end{bmatrix}
$$

#### 1. 求解特徵值 ($\lambda$)
計算特徵方程式 $\det(AA^T - \lambda I) = 0$：

$$
(11-\lambda)^2 - 1 = 0 \implies \lambda_1 = 12, \quad \lambda_2 = 10
$$

#### 2. 求解奇異值 ($\sigma$) 並構建 $\Sigma$
奇異值為特徵值的平方根 $\sigma_i = \sqrt{\lambda_i}$：
* $\sigma_1 = \sqrt{12} = 2\sqrt{3}$
* $\sigma_2 = \sqrt{10}$

矩陣 $\Sigma$ 為：

$$
\Sigma = \begin{bmatrix} \sqrt{12} & 0 & 0 \\ 0 & \sqrt{10} & 0 \end{bmatrix}
$$

#### 3. 求解左奇異向量 ($U$)
計算 $AA^T$ 的歸一化特徵向量：

* **對於 $\lambda_1 = 12$**:

    解方程
    

$$
\begin{bmatrix}
11-12 & 1\\
1 & 11-12
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
-1 & 1\\
1 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

==> x = y

    歸一化後：

$$
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
1/\sqrt{2} \\
1/\sqrt{2}
\end{bmatrix}$$



* **對於 $\lambda_2 = 10$**:
    解方程

$$
\begin{bmatrix}
11-10 & 1\\
1 & 11-10
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
1 & 1\\
1 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$



==> x = -y

    歸一化後：
    
$$
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
1/\sqrt{2} \\
-1/\sqrt{2}
\end{bmatrix}$$

矩陣 $U$ 為：

$$
U = \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \end{bmatrix}
$$

---

### 第二步：計算右奇異向量 ($V$)

利用公式 $\mathbf{v}_i = \frac{1}{\sigma_i} A^T \mathbf{u}_i$ 計算前兩個向量，第三個向量由零空間 (Null Space) 或外積求得。

#### 1. 計算 $\mathbf{v}_1$

$$
\mathbf{v}_1 = \frac{1}{\sqrt{12}} \begin{bmatrix} 3 & -1 \\ 1 & 3 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} 1/\sqrt{2} \\ 1/\sqrt{2} \end{bmatrix} = \frac{1}{\sqrt{6}} \begin{bmatrix} 1 \\ 2 \\ 1 \end{bmatrix}
$$

#### 2. 計算 $\mathbf{v}_2$

$$
\mathbf{v}_2 = \frac{1}{\sqrt{10}} \begin{bmatrix} 3 & -1 \\ 1 & 3 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} 1/\sqrt{2} \\ -1/\sqrt{2} \end{bmatrix} = \frac{1}{\sqrt{5}} \begin{bmatrix} 2 \\ -1 \\ 0 \end{bmatrix}
$$

#### 3. 計算 $\mathbf{v}_3$

尋找與 $\mathbf{v}_1, \mathbf{v}_2$ 正交的向量（解 $A\mathbf{x}=0$ 或計算外積）：

$$
\mathbf{v}_3 = \frac{1}{\sqrt{30}} \begin{bmatrix} 1 \\ 2 \\ -5 \end{bmatrix}
$$

矩陣 $V$ 為：

$$
V = \begin{bmatrix} \frac{1}{\sqrt{6}} & \frac{2}{\sqrt{5}} & \frac{1}{\sqrt{30}} \\ \frac{2}{\sqrt{6}} & -\frac{1}{\sqrt{5}} & \frac{2}{\sqrt{30}} \\ \frac{1}{\sqrt{6}} & 0 & -\frac{5}{\sqrt{30}} \end{bmatrix}
$$

---

## 3. 最終結果形式

$$
A = U \Sigma V^T
$$

$$
A = \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \end{bmatrix}
\begin{bmatrix} \sqrt{12} & 0 & 0 \\ 0 & \sqrt{10} & 0 \end{bmatrix}
\begin{bmatrix} \frac{1}{\sqrt{6}} & \frac{2}{\sqrt{6}} & \frac{1}{\sqrt{6}} \\ \frac{2}{\sqrt{5}} & -\frac{1}{\sqrt{5}} & 0 \\ \frac{1}{\sqrt{30}} & \frac{2}{\sqrt{30}} & -\frac{5}{\sqrt{30}} \end{bmatrix}
$$

---

## 4. 驗證正交性 (Verification)

正交矩陣需滿足：行向量長度為 1 且相互內積為 0。

### 驗證 $U$

$$
U = \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \end{bmatrix}
$$

1.  **長度檢查**：
    * $\|\mathbf{u}_1\|^2 = 1/2 + 1/2 = 1$
    * $\|\mathbf{u}_2\|^2 = 1/2 + 1/2 = 1$
2.  **正交檢查**：
    * $\mathbf{u}_1 \cdot \mathbf{u}_2 = 1/2 - 1/2 = 0$

**結論：$U$ 是正交矩陣。**

### 驗證 $V$

$$
V = \begin{bmatrix} \frac{1}{\sqrt{6}} & \frac{2}{\sqrt{5}} & \frac{1}{\sqrt{30}} \\ \frac{2}{\sqrt{6}} & -\frac{1}{\sqrt{5}} & \frac{2}{\sqrt{30}} \\ \frac{1}{\sqrt{6}} & 0 & -\frac{5}{\sqrt{30}} \end{bmatrix}
$$

1.  **長度檢查**：
    * $\|\mathbf{v}_1\|^2 = 1/6 + 4/6 + 1/6 = 1$
    * $\|\mathbf{v}_2\|^2 = 4/5 + 1/5 + 0 = 1$
    * $\|\mathbf{v}_3\|^2 = 1/30 + 4/30 + 25/30 = 1$
2.  **正交檢查**：
    * $\mathbf{v}_1 \cdot \mathbf{v}_2 = \frac{2}{\sqrt{30}} - \frac{2}{\sqrt{30}} + 0 = 0$
    * $\mathbf{v}_1 \cdot \mathbf{v}_3 = \frac{1+4-5}{\sqrt{180}} = 0$
    * $\mathbf{v}_2 \cdot \mathbf{v}_3 = \frac{2-2+0}{\sqrt{150}} = 0$

**結論：$V$ 是正交矩陣。**


```python
import numpy as np

# 1. 定義矩陣 A
A = np.array([[3, 1, 1],
              [-1, 3, 1]])

print("原始矩陣 A:")
print(A)
print("\n")

# 2. 使用 NumPy 執行 SVD
# np.linalg.svd 返回 U, s, Vh
# 其中 s 是一個包含奇異值的向量，Vh 是 V 的共軛轉置
U_np, s_np, Vh_np = np.linalg.svd(A)

# 將奇異值 s 轉換為對角矩陣 Sigma
Sigma_np = np.zeros(A.shape)
Sigma_np[:A.shape[0], :A.shape[0]] = np.diag(s_np)

# V 是 Vh 的轉置
V_np = Vh_np.T

print("NumPy 計算的 U 矩陣:")
print(U_np)
print("\n")

print("NumPy 計算的奇異值 (s):")
print(s_np)
print("\n")

print("NumPy 構建的 Sigma 矩陣:")
print(Sigma_np)
print("\n")

print("NumPy 計算的 V 矩陣:")
print(V_np)
print("\n")

# 3. 驗證 SVD: A = U @ Sigma @ V.T
A_reconstructed = U_np @ Sigma_np @ Vh_np
print("重建後的矩陣 A:")
print(A_reconstructed)
print("\n")

# 檢查重建矩陣與原始矩陣是否接近
print("原始矩陣 A 與重建矩陣 A_reconstructed 之間的差異 (應接近於 0):")
print(np.allclose(A, A_reconstructed))
print("\n")

# 4. 比較手動推導的結果與 NumPy 的結果 (提供參考)
print("手動推導的奇異值: sqrt(12) ≈ 3.464, sqrt(10) ≈ 3.162")
print("NumPy 計算的奇異值: ", s_np)

print("\n請注意，NumPy 計算的 U 和 V 矩陣的列向量可能與手動計算的結果在正負號上有所不同，但它們應該是等價的 (因為特徵向量的正負號是任意的，只要方向正確且歸一化即可)。")
print("例如，手動推導的 u1 = [1/sqrt(2), 1/sqrt(2)], 而 NumPy 可能會返回 [-1/sqrt(2), -1/sqrt(2)]。")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

# 1. 載入範例數據 (手寫數字圖片)
digits = load_digits()
X = digits.data
image = X[0].reshape(8, 8) # 取第一張圖

# 2. 執行 SVD 分解
# U: 左奇異向量, s: 奇異值, Vt: 右奇異向量
U, s, Vt = np.linalg.svd(image)

# 3. 建立對角矩陣 Sigma
Sigma = np.zeros((8, 8))
for i in range(8):
    Sigma[i, i] = s[i]

# 4. 重建圖像 (使用不同數量的奇異值)
# k=1: 只用最重要的 1 個特徵
k = 1
reconstructed_image = np.dot(U[:, :k], np.dot(Sigma[:k, :k], Vt[:k, :]))

print(f"原始奇異值: {s}")
print(f"使用前 {k} 個奇異值重建圖像完成")
```

```python
import matplotlib.pyplot as plt

# 繪製原始圖像與重建圖像
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('原始圖像')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(reconstructed_image, cmap='gray')
plt.title(f'重建圖像 (使用前 {k} 個奇異值)')
plt.axis('off')

plt.show()
```
